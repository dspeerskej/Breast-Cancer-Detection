```python
#replacing the values with nan values
cols = ['concavity_mean','concave points_mean','concavity_se','concave points_se',
        'concavity_worst','concave points_worst']
df[cols] = df[cols].replace({0.0:np.nan})
```

```python
#checking the nan values for the dataset
df.isna().sum().sort_values(ascending=False)[:7]
```

```python
#replacing nan with mean values
df.fillna(df.mean(),inplace=True)
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_Y=LabelEncoder()
df.iloc[:,1]=labelencoder_Y.fit_transform(df.iloc[:,1].values)
```

```python
#splitting the X and Y values
X = df.iloc[:,2:]
Y = df.iloc[:,1]

print(X.columns)
print('\n')
print(Y.name)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.25, random_state=0)
```

```python
#Standardising the dataset
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
def logreg (X_train, Y_train):
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression (random_state=0)
    log.fit(X_train, Y_train)
    print("Logistic Regression Training Accuracy:", log.score(X_train, Y_train))
    return log
```

```python
logrex=logreg(X_train, Y_train)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, logrex.predict(X_test))
print(cm)
```

```python
TP=cm[0][0]
TN=cm[1][1]
FN=cm[1][0]
FP=cm[0][1]
print("Testing accuracy of logistic regression model=", (TP+TN)/(TP+TN+FN+FP))
```

```python
#Trying out SVM
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(random_state=0)
svc.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
Y_pred = svc.predict(X_test)
print(f'Accuracy of test dataset: {accuracy_score(Y_test,Y_pred)}')
```

```python
#Trying RFC
from sklearn.ensemble import RandomForestClassifier as RFC
```

```python
rfc = RFC(random_state=0)
rfc.fit(X_train,Y_train)
```

```python
Y_pred = rfc.predict(X_test)
print(f'Accuracy of test dataset: {accuracy_score(Y_test,Y_pred)}')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(X.shape[1],activation='relu',input_dim = X.shape[1]))
model.add(Dense(500,activation='relu'))
model.add(Dense(500,activation='relu'))
model.add(Dense(500,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,Y_train,epochs=20,batch_size=128)
```

```python
model.summary()
```

```python
Y_pred = model.predict_classes(X_test)
```

```python
model.evaluate(X_test,Y_test.to_numpy())
```

```python
accuracy_score(Y_test,Y_pred)
```
